```python
import sys 
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('dataframe_train.csv')
df_train['action_type'] = np.array(df_train['action_type'] == 'DELIVERY').astype(int)
df_train
```

```python
df_train = df_train[['action_type', 'speed', 'max_load', 'weather_grade', 'source_type', 'grid_distance', 
                     'urgency', 'hour']]
df_train = pd.get_dummies(df_train)
for i in df_train['hour'].unique():
    df_train[str(i)] = (df_train['hour'] == i).astype(int)
df_train.info()
```

```python
x = df_train.drop(columns = ['action_type'])
y = df_train['action_type'].ravel()
x.info()
```

```python
#try to find the best parameters for the Logistic Regression
#l2 case
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression as LogitReg
from sklearn.model_selection import GridSearchCV
LR = LogitReg(penalty = 'l2', max_iter = 1000000)
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000] }
grid_1 = GridSearchCV(LR, param_grid, scoring = 'roc_auc', cv = 5)
grid_1.fit(x, y)

print("The best C for l2 is ",grid_1.best_params_) 
print() 
print("The AUC associated with the best C is %0.4f." % grid_1.best_score_)
```

```python
#l1 case
import statsmodels as sm
import sklearn
import scipy as sp
%matplotlib inline 
import matplotlib.pyplot as plt
import math

LR = LogitReg(penalty = 'l1', max_iter = 1000000, solver = 'saga')
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000] }

#Split the data into two parts because if we run all the original training set , it will take too much time. 
#And we compute the in-sample AUC below, just like what we did for l2.

X_train, X_test, y_train, y_test = tr_te_split(x, y, test_size=0.8)
grid_2 = GridSearchCV(LR, param_grid, scoring = 'roc_auc', cv = 5)
grid_2.fit(X_train, y_train)

print("The best C for l1 is ",grid_2.best_params_) 
print() 
print("The AUC associated with the best C is %0.4f." % grid_2.best_score_)
```

```python
#try KNN
from sklearn.model_selection import train_test_split as tr_te_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = tr_te_split(x, y, test_size=0.3)
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train),columns = X_train.columns)

param_grid = [{'n_neighbors':range(1,11)}]

clf = KNeighborsClassifier()

# The evaluation method is AUC.
grid_1 = GridSearchCV(clf, param_grid, scoring = 'roc_auc',cv=5)

grid_1.fit(X_train,y_train)
print("The best k is ",grid_1.best_params_) 
print("\n") 
print("The AUC associated with the best k is %0.4f." % grid_1.best_score_)
```

```python
#try random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

x = df_train.drop(columns = ['action_type'])
y = df_train['action_type'].ravel()

tree2 = RandomForestClassifier()
param_grid = {"max_depth": [1, 3, 5, 7, 9],
              "ccp_alpha": np.arange(0, 0.01, 0.005).tolist(),
              "min_samples_split": [10, 20, 40],
              "min_samples_leaf": [20, 40, 60],
              "max_leaf_nodes": [5, 20, 80]}

grid_cv_tree2 = GridSearchCV(tree2, param_grid, cv = 5)
grid_cv_tree2.fit(x,y)

print("The best parameters are ",grid_cv_tree2.best_params_) 
print() 
print("The score is %0.4f." % grid_cv_tree2.best_score_)
```

```python
#try decision tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import GridSearchCV
tree = DecisionTreeClassifier() 
param_grid = {"max_depth": [2, 3, 4],
              "ccp_alpha": np.arange(0, 0.01, 0.001).tolist(),
              "min_samples_split": [10, 20, 40],
              "min_samples_leaf": [20, 40, 60],
              "max_leaf_nodes": [5, 20, 80]}
              
grid_cv_tree = GridSearchCV(tree, param_grid, scoring = 'roc_auc', cv = 5)
grid_cv_tree.fit(x, y)

print("The best parameters are ",grid_cv_tree.best_params_) 
print() 
print("The score is %0.4f." % grid_cv_tree.best_score_)
```

```python
#So we fit a decision tree model on the entire training set, and evaluate the entire model accuracy on the testing set.
from sklearn.model_selection import train_test_split as tr_te_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = tr_te_split(x, y, test_size=0.3)
scaler = StandardScaler().fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)

tree = DecisionTreeClassifier(ccp_alpha = 0, max_depth = 4, max_leaf_nodes=20, min_samples_leaf=20, min_samples_split=10).fit(X_train_st,y_train)
```

```python
def accuracy(tn, fp, fn, tp):
    return (tn+tp)/(tn+fp+fn+tp)

def printCM(tn, fp, fn, tp):
    print('{: <9} {: <9} {: <9}'.format(' ',' ','predicted'))
    print('         --------------------')
    print('{: <9}|{: <9d}|{: <9d}'.format('Actual',0,1))
    print('-----------------------------')
    print('{: <9d}|tn={: <6d}|fp={: <9d}'.format(0,tn,fp))
    print('-----------------------------')
    print('{: <9d}|fn={: <6d}|tp={: <9d}'.format(1,fn,tp))
    
from sklearn.metrics import confusion_matrix
pred = tree.predict(X_test_st)
tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
print('The testing accuracy is {:.4f}.'.format(accuracy(tn, fp, fn, tp)))
print('\n')
printCM(tn, fp, fn, tp)

F1 = 0
def Mean_F1_Score(tp,fp,fn):
    p = tp/(tp+fp)
    r = tp/(tp+fn)
    F1 = 2*p*r/(p+r)
    return F1
print('\n')
print('The mean-f1-score is {:.4f}.'.format(Mean_F1_Score(tp,fp,fn)))
```
